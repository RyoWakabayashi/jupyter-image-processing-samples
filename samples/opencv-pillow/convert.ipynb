# OpenCV Pillow 変換

```python
from PIL import Image, ImageOps
from matplotlib import pyplot as plt
import cv2
import numpy as np
```

## OpenCV to Pillow

```python
img_cv2 = cv2.imread('../lena.jpg')
type(img_cv2)
```

```python
plt.imshow(cv2.cvtColor(img_cv2, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
imp_pil = Image.fromarray(
    cv2.cvtColor(img_cv2, cv2.COLOR_BGR2RGB)
)
type(imp_pil)
```

```python
plt.imshow(np.asarray(imp_pil))
plt.show()
```

## Pillow to OpenCV

```python
img_pil = Image.open('../lena.jpg')
type(imp_pil)
```

```python
plt.imshow(np.asarray(imp_pil))
plt.show()
```

```python
img_cv2 = cv2.cvtColor(np.array(imp_pil, dtype=np.uint8), cv2.COLOR_RGB2BGR)
type(img_cv2)
```

```python
plt.imshow(cv2.cvtColor(img_cv2, cv2.COLOR_BGR2RGB))
plt.show()
```
