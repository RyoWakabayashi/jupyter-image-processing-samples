# OpenCV による画像の回転・反転

```python
import cv2
from matplotlib import pyplot as plt
```

## 元画像

```python
img_org = cv2.imread('../lena.jpg')
img_org = cv2.cvtColor(img_org, cv2.COLOR_BGR2RGB)
plt.imshow(img_org)
plt.show()
```

## 画像の回転

### 時計回りに90度

```python
img = cv2.rotate(img_org, cv2.ROTATE_90_CLOCKWISE)
plt.imshow(img)
plt.show()
```

### 反時計回りに９０度

```python
img = cv2.rotate(img_org, cv2.ROTATE_90_COUNTERCLOCKWISE)
plt.imshow(img)
plt.show()
```

### 180度

```python
img = cv2.rotate(img_org, cv2.ROTATE_180)
plt.imshow(img)
plt.show()
```

### 任意の角度

```python
# 画像の高さと幅を取得（shape = 高さ,幅,RGB）
height, width = img_org.shape[:2]
# 中心座標を取得
center = (int(width/2), int(height/2))

# 反時計回りの角度
angle = 45.0
# 拡大縮小率
scale = 1.0
# 背景色（指定しなければ黒）
bg_color = (255, 255, 255)

mat = cv2.getRotationMatrix2D(center, angle , scale)
img = cv2.warpAffine(img_org, mat, (width, height), borderValue=bg_color)
plt.imshow(img)
plt.show()
```

## 画像の反転

### 左右反転

```python
img = cv2.flip(img_org, 1)
plt.imshow(img)
plt.show()
```

### 上下反転

```python
img = cv2.flip(img_org, 0)
plt.imshow(img)
plt.show()
```

### 上下左右反転

```python
img = cv2.flip(img_org, -1)
plt.imshow(img)
plt.show()
```
