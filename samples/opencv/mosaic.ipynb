# OpenCVによるモザイク加工

```python
import cv2
from matplotlib import pyplot as plt
```

## 元画像

```python
img_org = cv2.imread('../lena.jpg')
img_org = cv2.cvtColor(img_org, cv2.COLOR_BGR2RGB)
plt.imshow(img_org)
plt.show()
```

## 全体モザイク

```python
# 元画像の高さと幅を取得（shape = 高さ,幅,RGB）
height, width = img_org.shape[:2]

# 0.1倍に縮小する
img = cv2.resize(
    img_org, None, fx=0.1, fy=0.1,
    interpolation=cv2.INTER_NEAREST
)
# 元の大きさに戻す
img = cv2.resize(
    img, (width, height),
    interpolation=cv2.INTER_NEAREST
)

plt.imshow(img)
plt.show()
```

## 部分モザイク

```python
# 顔の部分を取得
face_org = img_org[75:150, 150:250, ]

# 顔の高さと幅を取得（shape = 高さ,幅,RGB）
face_height, face_width = face_org.shape[:2]

# 0.1倍に縮小する
face = cv2.resize(
    face_org, None, fx=0.1, fy=0.1,
    interpolation=cv2.INTER_NEAREST
)
# 元の大きさに戻す
face = cv2.resize(
    face, (face_width, face_height),
    interpolation=cv2.INTER_NEAREST
)

# 顔の位置にモザイク化したものを貼り付ける
img = img_org.copy()
img[75:150, 150:250, ] = face

plt.imshow(img)
plt.show()
```
