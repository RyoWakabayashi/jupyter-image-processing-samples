# OpenCV による画像の読込・書込

```python
import cv2
from matplotlib import pyplot as plt
```

## 画像の読込

```python
img = cv2.imread('../lena.jpg')
```

## matplotlib で表示するため、BGRからRGBに変換する

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

## 画像の表示

```python
plt.imshow(img)
plt.show()
```

## 画像の反転

```python
img = cv2.flip(img, 1)
plt.imshow(img)
plt.show()
```

## 保存前にRGBからGBRに戻す

```python
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
```

## 画像の保存

```python
cv2.imwrite('../lena_flip.jpg', img)
```
