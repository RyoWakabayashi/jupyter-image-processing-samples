# OpenCV による画像の結合・合成

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

## 元画像

```python
img_org = cv2.imread('../lena.jpg')
img_org = cv2.cvtColor(img_org, cv2.COLOR_BGR2RGB)
plt.imshow(img_org)
plt.show()
```

## 画像の結合

### 縦に結合

```python
img = cv2.vconcat([img_org, img_org])
plt.imshow(img)
plt.show()
```

### 横に結合

```python
img = cv2.hconcat([img_org, img_org])
plt.imshow(img)
plt.show()
```

## 画像の合成

### 背景画像

```python
# 元画像の高さと幅を取得（shape = 高さ,幅,RGB）
height, width = img_org.shape[:2]

# 元画像より前後左右に30px大きい背景画像を作る
margin = 30

img_bg = np.ones((height + 2 * margin, width + 2 * margin, 3), dtype=int) * 255
plt.imshow(img_bg)
plt.show()
```

```python
# 背景画像をコピー
img = img_bg.copy()

# 指定した範囲に貼り付け
img[:-60, :-60, ] = img_org

plt.imshow(img)
plt.show()
```

```python
# 背景画像をコピー
img = img_bg.copy()

# 指定した範囲に貼り付け
img[30:-30, 30:-30, ] = img_org

plt.imshow(img)
plt.show()
```

```python
# 背景画像をコピー
img = img_bg.copy()

# 指定した範囲に貼り付け
img[60:, 60:, ] = img_org

plt.imshow(img)
plt.show()
```

```python
# 背景画像をコピー
img = img_bg.copy()

# 指定した範囲に貼り付け
img[90:, 90:, ] = img_org[:-30, :-30, ]

plt.imshow(img)
plt.show()
```
