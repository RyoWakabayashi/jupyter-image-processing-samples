# Pillow による画像の回転・反転

```python
from PIL import Image, ImageOps
from matplotlib import pyplot as plt
import numpy as np
```

## 元画像

```python
img_org = Image.open('../lena.jpg')
plt.imshow(np.asarray(img_org))
plt.show()
```

## 画像の回転

### 時計回りに90度

```python
img = img_org.rotate(-90, expand=True)
plt.imshow(np.asarray(img))
plt.show()
```

### 反時計回りに９０度

```python
img = img_org.rotate(90, expand=True)
plt.imshow(np.asarray(img))
plt.show()
```

### 180度

```python
img = img_org.rotate(180)
plt.imshow(np.asarray(img))
plt.show()
```

### 任意の角度

```python
img = img_org.rotate(45, expand=True)
plt.imshow(np.asarray(img))
plt.show()
```

### 任意の角度で任意の背景色

```python
img = img_org.convert('RGBA')
img = img.rotate(45, expand=True)

# 背景用画像
bg_color = (255, 0, 0)
bg_img = Image.new('RGBA', img.size, bg_color)

# 背景用画像と合成
img = Image.composite(img, bg_img, img)

plt.imshow(np.asarray(img))
plt.show()
```

## 画像の反転

### 左右反転

```python
img = ImageOps.mirror(img_org)
plt.imshow(np.asarray(img))
plt.show()
```

### 上下反転

```python
img = ImageOps.flip(img_org)
plt.imshow(np.asarray(img))
plt.show()
```

### 上下左右反転

```python
img = ImageOps.flip(ImageOps.mirror(img_org))
plt.imshow(np.asarray(img))
plt.show()
```
